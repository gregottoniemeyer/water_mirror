```python
import pandas as pd
import os
from bs4 import BeautifulSoup

# Base folder path
folder_path = "./ncedc_html_files/"

# List of filenames with folder path
filenames = [
    os.path.join(folder_path, filename)
    for filename in [
        "nc74033306_MT.html", "nc74030356_MT.html", "nc74029551_MT.html",
        "nc74018946_MT.html", "nc74009641_MT.html", "nc74005576_MT.html",
        "nc74001951_MT.html", "nc74000441_MT.html", "nc73999211_MT.html",
        "nc73998256_MT.html", "nc73997816_MT.html", "nc73994586_MT.html",
        "nc73992726_MT.html", "nc73975501_MT.html", "nc73961261_MT.html",
        "nc73952605_MT.html", "nc73952475_MT.html", "nc73948665_MT.html",
        "nc73947835_MT.html", "nc73947830_MT.html", "nc73947486_MT.html",
        "nc73944271_MT.html", "nc73944166_MT.html", "nc73943846_MT.html",
        "nc73943841_MT.html", "nc73943821_MT.html", "nc73938736_MT.html",
        "nc73938706_MT.html", "nc73938626_MT.html", "nc73934641_MT.html",
        "nc73928421_MT.html", "nc73925281_MT.html", "nc73923661_MT.html",
        "nc73922596_MT.html", "nc73918291_MT.html", "nc73909066_MT.html",
        "nc73908746_MT.html", "nc73908361_MT.html", "nc71133194_MT.html"
    ]
]

# Initialize a list to store data for all files
all_data = []

# Function to process each file
def process_file(filename):
    try:
        with open(filename, "r") as file:
            soup = BeautifulSoup(file, "html.parser")
            
            # Locate Origin Time
            origin_time_tag = soup.find(string="Origin Time")
            if not origin_time_tag:
                print(f"Origin Time not found in file: {filename}")
                return
            timestamp = origin_time_tag.find_next().text.strip()
            
            # Locate data rows for Axis, Value, Plunge, Azimuth
            for table in soup.find_all("table"):  # Assuming data is in tables
                for row in table.find_all("tr"):
                    columns = row.find_all("td")
                    if len(columns) == 4:
                        axis = columns[0].text.strip()
                        if axis in ["T", "N", "P"]:  # Process T, N, P rows only
                            value = float(columns[1].text.strip())
                            plunge = float(columns[2].text.strip())
                            azimuth = float(columns[3].text.strip())
                            all_data.append({
                                "Timestamp": timestamp,
                                "Axis": axis,
                                "Value": value,
                                "Plunge": plunge,
                                "Azimuth": azimuth
                            })
    except Exception as e:
        print(f"Error processing file {filename}: {e}")

# Process each file
for filename in filenames:
    process_file(filename)

# Convert the data to a DataFrame
df = pd.DataFrame(all_data)
```

```python
df = df.sort_values(by="Timestamp")
df.set_index("Timestamp", inplace=True)
```

```python
df.head()
```

```python
# Pivot the table
pivot_df = df.pivot(columns='Axis', values=['Value', 'Plunge', 'Azimuth'])

# Flatten the MultiIndex columns
pivot_df.columns = [f"{axis}_{param}" for param, axis in pivot_df.columns]

# Reset the index (optional, if you want the timestamp as a column again)
pivot_df.reset_index(inplace=True)

# Display the result
pivot_df.tail(25)
```

```python
import matplotlib.pyplot as plt

# Define metrics and axes for visualization
metrics = ["Value", "Plunge", "Azimuth"]
axes = ["P", "N", "T"]

# Iterate through metrics and plot for each axis
for metric in metrics:
    plt.figure(figsize=(10, 6))
    for axis in axes:
        column = f"{axis}_{metric}"
        if column in pivot_df:
            plt.plot(pivot_df.index, pivot_df[column], label=f"{axis} - {metric}")

    # Customize the plot
    plt.title(f"Trends for {metric}", fontsize=14)
    plt.xlabel("Timestamp", fontsize=12)
    plt.ylabel(metric, fontsize=12)
    plt.legend()
    plt.grid(True)
    plt.tight_layout()

    # Show the plot
    plt.show()
```
