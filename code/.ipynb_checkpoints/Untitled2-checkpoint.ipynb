```python
import os
import subprocess

blend_file = "../CGI/fourSeasons_26.blend"
frames_per_slice = 1000
total_frames = 35066

commands = []
for start in range(0, total_frames, frames_per_slice):
    end = min(start + frames_per_slice - 1, total_frames)
    command = f"blender -b {blend_file} -s {start} -e {end} -a"
    commands.append(command)

# Launch commands in parallel
processes = [subprocess.Popen(cmd, shell=True) for cmd in commands]

# Wait for all processes to finish
for p in processes:
    p.wait()
```
